Graph for healthy and lifestyle dataset:



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(42)
data = {
    "Age": np.random.randint(18, 65, 100),
    "Exercise_Hours_Per_Week": np.random.uniform(0, 10, 100),
    "Sleep_Hours_Per_Night": np.random.uniform(4, 10, 100),
    "Healthy_Meals_Per_Week": np.random.randint(0, 21, 100),
    "BMI": np.random.uniform(18, 35, 100),
    "Stress_Level": np.random.uniform(1, 10, 100)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Add a derived health indicator for simplicity
df["Overall_Health_Score"] = (
    0.4 * df["Exercise_Hours_Per_Week"] +
    0.3 * (10 - df["Stress_Level"]) +
    0.3 * df["Healthy_Meals_Per_Week"] / 20
)

# Display the first few rows of the data
print(df.head())

# Visualize relationships between variables
sns.pairplot(df, diag_kind="kde")
plt.suptitle("Health and Lifestyle Relationships", y=1.02)
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Health and Lifestyle Factors")
plt.show()

# Analyze the strongest correlations
correlation_matrix = df.corr()
health_correlations = correlation_matrix["Overall_Health_Score"].sort_values(ascending=False)
print("Correlations with Overall Health Score:")
print(health_correlations)
```

Box plot,Violin plot and Swarm plot for the following dataset:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(42)
data = {
    "Age": np.random.randint(18, 65, 100),
    "Exercise_Hours_Per_Week": np.random.uniform(0, 10, 100),
    "Sleep_Hours_Per_Night": np.random.uniform(4, 10, 100),
    "Healthy_Meals_Per_Week": np.random.randint(0, 21, 100),
    "BMI": np.random.uniform(18, 35, 100),
    "Stress_Level": np.random.uniform(1, 10, 100)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Add categorical health level based on BMI and stress
df["Health_Level"] = pd.cut(
    df["BMI"],
    bins=[0, 18.5, 24.9, 29.9, 100],
    labels=["Underweight", "Normal", "Overweight", "Obese"]
)

# Boxplot: Exercise Hours vs. Health Level
plt.figure(figsize=(8, 6))
sns.boxplot(x="Health_Level", y="Exercise_Hours_Per_Week", data=df, palette="Set2")
plt.title("Boxplot: Exercise Hours vs. Health Level")
plt.xlabel("Health Level")
plt.ylabel("Exercise Hours Per Week")
plt.show()

# Violin plot: Sleep Hours vs. Health Level
plt.figure(figsize=(8, 6))
sns.violinplot(x="Health_Level", y="Sleep_Hours_Per_Night", data=df, palette="muted")
plt.title("Violin Plot: Sleep Hours vs. Health Level")
plt.xlabel("Health Level")
plt.ylabel("Sleep Hours Per Night")
plt.show()

# Swarm plot: Healthy Meals vs. Stress Level
plt.figure(figsize=(8, 6))
sns.swarmplot(x="Health_Level", y="Healthy_Meals_Per_Week", data=df, palette="cool")
plt.title("Swarm Plot: Healthy Meals vs. Health Level")
plt.xlabel("Health Level")
plt.ylabel("Healthy Meals Per Week")
plt.show()
```

KDE and Run plot for the following dataset:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(42)
data = {
    "Age": np.random.randint(18, 65, 100),
    "Exercise_Hours_Per_Week": np.random.uniform(0, 10, 100),
    "Sleep_Hours_Per_Night": np.random.uniform(4, 10, 100),
    "Healthy_Meals_Per_Week": np.random.randint(0, 21, 100),
    "BMI": np.random.uniform(18, 35, 100),
    "Stress_Level": np.random.uniform(1, 10, 100)
}

# Create a DataFrame
df = pd.DataFrame(data)

# KDE Plot: Stress Level vs. Exercise Hours
plt.figure(figsize=(8, 6))
sns.kdeplot(
    x=df["Exercise_Hours_Per_Week"],
    y=df["Stress_Level"],
    cmap="Blues",
    fill=True,
    cbar=True
)
plt.title("KDE Plot: Stress Level vs. Exercise Hours")
plt.xlabel("Exercise Hours Per Week")
plt.ylabel("Stress Level")
plt.show()

# Rug Plot: Sleep Hours
plt.figure(figsize=(8, 6))
sns.kdeplot(df["Sleep_Hours_Per_Night"], fill=True, color="green", label="KDE")
sns.rugplot(df["Sleep_Hours_Per_Night"], color="black", height=0.1)
plt.title("KDE and Rug Plot: Sleep Hours Per Night")
plt.xlabel("Sleep Hours Per Night")
plt.ylabel("Density")
plt.legend()
plt.show()
```

Piechart for the following dataset:

```python
import matplotlib.pyplot as plt

# Example dataset: Proportions of lifestyle factors
lifestyle_data = {
    'Exercise': 20,
    'Diet': 25,
    'Sleep': 30,
    'Screen Time': 15,
    'Other Activities': 10
}

# Labels and values
labels = list(lifestyle_data.keys())
values = list(lifestyle_data.values())

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)

# Adding a title
plt.title('Health and Lifestyle Proportions')
plt.show()
```
